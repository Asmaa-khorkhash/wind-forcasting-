```python
df.drop(columns = ['WTG','ControlBoxTemperature','Unnamed: 0'], inplace = True) #We drop WTG because every value was the same
```

```python
columns = ['AmbientTemperatue' ,
           'BearingShaftTemperature',
           'Blade1PitchAngle', 
           'Blade2PitchAngle' ,
           'Blade3PitchAngle',
           'GearboxBearingTemperature',
           'GearboxOilTemperature',
           'GeneratorRPM','GeneratorWinding1Temperature','GeneratorWinding2Temperature',
           'HubTemperature',
           'MainBoxTemperature',
           'NacellePosition',
           'ReactivePower',
           'RotorRPM',
           'TurbineStatus', 
           'WindDirection',
           'WindSpeed']
for n in columns:
    df[n].fillna(df[n].median(),inplace = True)
for d in df:
    df.dropna(axis=0,how = 'any',inplace = True)
```

```python
# Data scaling

df_scaler = MinMaxScaler().fit(df)
df_scaled = pd.DataFrame(df_scaler.transform(df))

x_scaled = df_scaled.iloc[:, 1:]
y_scaled = df_scaled.iloc[:, 0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.3, random_state=1)
```

```python
performer_model = Performer.Performer_model(1000,18,1,6,4,2,5).getModel()
performer_model.compile(optimizer="adam" , loss ="mse")
```

```python
performer_model.fit(x_train, y_train, epochs = 20 ,batch_size = 32, verbose= 1)
```

```python
Evaluation_Metrix.print_metrics(y_train, performer_model.predict(x_train), y_test, performer_model.predict(x_test))
```
